```python
"""
-----------------------------------------------------------------------------
Bays decision rule demo
-----------------------------------------------------------------------------
AUTHOR: Soumitra Samanta (soumitra.samanta@gm.rkmvu.ac.in)
-----------------------------------------------------------------------------
"""

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
apple_prior = 0.9
orange_prior = 1 - apple_prior

# Plot priors
plt.bar(['Apple', 'Orange'], [apple_prior, orange_prior], color=['red', 'orange'])
plt.title('Prior')
plt.show()
```

```python
def gaussian_distribution(x, mu, sigma):
    
    """One dimensional Gaussian pdf for given x, mu and sigma"""
    
    return norm(mu, sigma).pdf(x)

x = np.arange(0, 1000, 0.1)

# Apple
apple_mu = 400
apple_sigma = 100
apple_class_conditional_dist = gaussian_distribution(x, apple_mu, apple_sigma)

orange_mu = 150
orange_sigma = 50
orange_class_conditional_dist = gaussian_distribution(x, orange_mu, orange_sigma)

# Plot class conditional distribution
plt.plot(x, apple_class_conditional_dist, label='Apple', color='r')
plt.plot(x, orange_class_conditional_dist, label='Orange', color='orange')
plt.xlabel('weight')
plt.ylabel('$P(weight/c_i)$')
plt.grid(linestyle='--')
plt.xticks(np.arange(x.min(), x.max()+1, step=100))
plt.legend()
plt.title('Class conditional distribution')
plt.show()


    
```

```python
# Priors
apple_prior = 0.1
orange_prior = 1 - apple_prior

# Likelihood
x = np.arange(0, 1000, 0.1)
# Apple
apple_mu = 400
apple_sigma = 100
apple_class_conditional_dist = gaussian_distribution(x, apple_mu, apple_sigma)
# Orange
orange_mu = 150
orange_sigma = 50
orange_class_conditional_dist = gaussian_distribution(x, orange_mu, orange_sigma)

# Evidence
evidence = (apple_prior*apple_class_conditional_dist) + (orange_prior*orange_class_conditional_dist)

# Posteriors
apple_posterior = (apple_class_conditional_dist * apple_prior) / evidence
orange_posterior = (orange_class_conditional_dist * orange_prior) / evidence

# apple_posterior = apple_class_conditional_dist * apple_prior
# orange_posterior = orange_class_conditional_dist * orange_prior

plt.figure(figsize=(10, 3))

# Plot Prior
plt.subplot(121)
plt.bar(['Apple', 'Orange'], [apple_prior, orange_prior], color=['r', 'orange'])
plt.grid(linestyle='--')
plt.title('Prior')

# Plot class conditional distribution
plt.subplot(122)
plt.subplots_adjust(wspace=0.3)
plt.plot(x, apple_class_conditional_dist, label='Apple', color='r')
plt.plot(x, orange_class_conditional_dist, label='Orange', color='orange')
plt.xlabel('weight')
plt.ylabel('$P(weight/c_i)$')
plt.grid(linestyle='--')
plt.xticks(np.arange(x.min(), x.max()+1, step=100))
plt.legend()
plt.title('Class conditional distribution')
plt.show()

# Plot posterior distribution
# plt.subplot(133)
# plt.subplots_adjust(wspace=0.5)
plt.figure(figsize=(5, 3))
plt.plot(x, apple_posterior, label='Apple', color='r')
plt.plot(x, orange_posterior, label='Orange', color='orange')
plt.xlabel('weight')
plt.ylabel('$P(c_i/weight)$')
plt.grid(linestyle='--')
plt.xticks(np.arange(x.min(), x.max()+1, step=100))
plt.legend()
plt.title('Posterior distribution')
plt.show()

```
